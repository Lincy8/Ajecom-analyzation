```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt,seaborn as sns
%matplotlib inline
```

```python
#file reading
ecom = pd.read_csv('AJEcom.csv')
ecom.drop(ecom.columns[ecom.columns.str.contains('unnamed',case=False)],axis =1,inplace =True)
ecom.head()
```

```python
#Total rows and columns 
ecom.shape
```

```python
#datatypes
ecom.dtypes
```

```python
#descriotive statistic
ecom.describe()
```

```python
#All columns name
ecom.columns
```

```python
#finding anu=y null values present
ecom.isnull().sum()
```

```python
#extracting data

ecom['tran_date'] = pd.to_datetime(ecom['tran_date'])
ecom['tran_date']
```

```python
#year
ecom['tran_date'].dt.year
ecom
```

```python
ecom['year'] =ecom['tran_date'].dt.year
ecom
```

# Yearly-wise Trend chart

```python
ecom_trend = ecom.groupby('year').sum()['Total _amount'].reset_index()
```

```python
plt.figure(figsize=[15,5])
plt.plot(ecom_trend['year'],ecom_trend['Total _amount'])
plt.xticks(rotation = 'vertical',size=8)

plt.title('Yearly-wise Trend Chart')
plt.ylabel('Tota_amount')
plt.xlabel('year');
```

# High demand and low demand products

```python
product_sold=pd.DataFrame(ecom.groupby('Product_cat_name').sum()['Quantity'])
```

```python
product_sold.sort_values('Quantity',ascending=False)
```

# which is the top  product sold

```python
ecom.groupby('Product_cat_name')['Quantity'].sum().plot(kind='bar', color = 'peru');
plt.title('Toltal sold products')
plt.ylabel('Quantity')
plt.xlabel('Product_name');
```

# Gender-wise maximum purchase done 

```python
ecom['Gender'].value_counts().plot.pie(figsize=(6, 6), autopct='%.2f')
plt.title('Gender-wise purchase ratio')
plt.xlabel('Category of gender');
```

# Which store have made the maxium profit

```python
ecom.groupby('Store_type')['Quantity'].sum().plot(kind='bar', color = 'red');
plt.title('Store-wise Profit')
plt.ylabel('Quantity')
plt.xlabel('store_type');
```

# Which age group of people have done the most purchase

```python
#yearextract DOB
ecom['DOB'] = pd.to_datetime(ecom['DOB'])
ecom['DOB']
```

```python
ecom['year_DOB'] =ecom['DOB'].dt.year
ecom
```

```python
ecom_age = ecom.groupby('year_DOB').count()['Total _amount'].reset_index()
```

```python
plt.figure(figsize=[15,5])
plt.plot(ecom_age['year_DOB'],ecom_age['Total _amount'])
plt.xticks(rotation = 'vertical',size=8)

plt.title('age-wise max spend')
plt.ylabel('Tota_amount')
plt.xlabel('year_DOB')
```

# age-wise-count

```python
ecom.year_DOB.hist(bins=8,color='green');
plt.xticks(rotation = 'vertical',size=8)
plt.title("year_DOB")
plt.ylabel('Total_count_of_age_group')
plt.xlabel('list_of_age');
```
